```python
import matplotlib.pyplot as plt

# Dữ liệu
values = [71.3, 95.73, 100, 97.73]
labels = ['001', '002', '003', '004']

# Tạo biểu đồ
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(labels, values, color='orange', width=0.5)  # Kích thước cột nhỏ hơn với width=0.5

# Thiết lập nhãn trục và tiêu đề
ax.set_xlabel('Nhóm thiết bị', fontsize=12)
ax.set_ylabel('%', fontsize=12)
ax.set_title('Đồ thị phụ tải về công suất cho mỗi nhóm máy', fontsize=14)

# Hiển thị giá trị tại đỉnh cột
for bar in bars:
    height = bar.get_height()
    ax.annotate('{:.1f}'.format(height),
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom')

# Thêm đường kẻ
ax.grid(True, linestyle='--', linewidth=0.5, color='gray')
ax.tick_params(axis='both', which='major', labelsize=10)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
